```python
import pygame
import sys

# Инициализация Pygame
pygame.init()

# Константы
WIDTH, HEIGHT = 800, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
GREEN = (0, 255, 0)

# Настройки экрана
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Paint Tool")
screen.fill(WHITE)

# Переменные
clock = pygame.time.Clock()
drawing = False
last_pos = None
mode = "pen"  # Режимы: 'pen', 'rect', 'circle', 'eraser'
color = BLACK
radius = 10  # Размер кисти или ластика

# Основной цикл
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False
        
        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_r:
                mode = "rect"  # Рисование прямоугольников
            elif event.key == pygame.K_c:
                mode = "circle"  # Рисование кругов
            elif event.key == pygame.K_p:
                mode = "pen"  # Режим кисти
            elif event.key == pygame.K_e:
                mode = "eraser"  # Ластик
            elif event.key == pygame.K_1:
                color = BLACK  # Выбор черного цвета
            elif event.key == pygame.K_2:
                color = RED  # Выбор красного цвета
            elif event.key == pygame.K_3:
                color = GREEN  # Выбор зеленого цвета
            elif event.key == pygame.K_4:
                color = BLUE  # Выбор синего цвета

        elif event.type == pygame.MOUSEBUTTONDOWN:
            drawing = True
            last_pos = event.pos

        elif event.type == pygame.MOUSEBUTTONUP:
            drawing = False
            if mode == "rect":
                pygame.draw.rect(screen, color, pygame.Rect(min(last_pos[0], event.pos[0]),
                                                             min(last_pos[1], event.pos[1]),
                                                             abs(event.pos[0] - last_pos[0]),
                                                             abs(event.pos[1] - last_pos[1])), 2)
            elif mode == "circle":
                radius = max(abs(event.pos[0] - last_pos[0]), abs(event.pos[1] - last_pos[1])) // 2
                pygame.draw.circle(screen, color, last_pos, radius, 2)

        elif event.type == pygame.MOUSEMOTION and drawing:
            if mode == "pen":
                pygame.draw.line(screen, color, last_pos, event.pos, radius)
                last_pos = event.pos
            elif mode == "eraser":
                pygame.draw.line(screen, WHITE, last_pos, event.pos, radius)
                last_pos = event.pos
    
    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```

Рисование кистью – P

Прямоугольник – R (зажать и тянуть)

Круг – C (зажать и тянуть)

Ластик – E

Выбор цвета:

1 – Черный

2 – Красный

3 – Зеленый

4 – Синий
