```python
import pygame
import sys
import random
from pygame.locals import *


pygame.init()


SCREEN_WIDTH = 400
SCREEN_HEIGHT = 600
DISPLAYSURF = pygame.display.set_mode((SCREEN_WIDTH, SCREEN_HEIGHT))
pygame.display.set_caption("Car Game with Coins")


FPS = 60
FramePerSec = pygame.time.Clock()

WHITE = (255, 255, 255)
BLACK = (0, 0, 0)

# Загрузка изображений
PLAYER_IMG = pygame.image.load("Player.jpg")
ENEMY_IMG = pygame.image.load("Enemy.png")
COIN_IMG = pygame.image.load("coin.jpg")
COIN_IMG = pygame.transform.scale(COIN_IMG, (COIN_IMG.get_width() // 2, COIN_IMG.get_height() // 2))


class Enemy(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = ENEMY_IMG
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)
        self.speed = 5  # Начальная скорость

    def move(self):
        self.rect.move_ip(0, self.speed)
        if self.rect.top > SCREEN_HEIGHT:
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)

    def draw(self, surface):
        surface.blit(self.image, self.rect)

        
class Player(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = PLAYER_IMG
        self.rect = self.image.get_rect()
        self.rect.center = (160, 520)

    def update(self):
        pressed_keys = pygame.key.get_pressed()
        if self.rect.left > 0 and pressed_keys[K_LEFT]:
            self.rect.move_ip(-5, 0)
        if self.rect.right < SCREEN_WIDTH and pressed_keys[K_RIGHT]:
            self.rect.move_ip(5, 0)

    def draw(self, surface):
        surface.blit(self.image, self.rect)

class Coin(pygame.sprite.Sprite):
    def __init__(self):
        super().__init__()
        self.image = COIN_IMG
        self.rect = self.image.get_rect()
        self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), random.randint(0, SCREEN_HEIGHT // 2))
        self.value = random.choice([1, 2, 5])  # Случайный вес монеты

    def move(self):
        self.rect.move_ip(0, 3)
        if self.rect.top > SCREEN_HEIGHT:
            self.rect.top = 0
            self.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)
            self.value = random.choice([1, 2, 5])  # Обновляем вес монеты при новом появлении

    def draw(self, surface):
        surface.blit(self.image, self.rect)

# Создание объектов
P1 = Player()
E1 = Enemy()
coins = pygame.sprite.Group()
for _ in range(5):  # Количество монет на экране
    coins.add(Coin())

# Счетчик монет
score = 0
font = pygame.font.Font(None, 36)

# Уровень сложности (каждые 10 очков ускоряем врага)
SCORE_TO_INCREASE_SPEED = 10

# Основной игровой цикл
while True:
    for event in pygame.event.get():
        if event.type == QUIT:
            pygame.quit()
            sys.exit()

    # Обновление объектов
    P1.update()
    E1.move()
    for coin in coins:
        coin.move()
        if P1.rect.colliderect(coin.rect):  # Проверка столкновения с монетой
            score += coin.value  # Добавляем вес монеты к счету
            coin.rect.top = 0
            coin.rect.center = (random.randint(40, SCREEN_WIDTH - 40), 0)
            coin.value = random.choice([1, 2, 5])  # Случайное значение для следующей монеты

            # Увеличиваем скорость противника каждые 10 очков
            if score % SCORE_TO_INCREASE_SPEED == 0:
                E1.speed += 1  # Увеличиваем скорость

    # Отрисовка
    DISPLAYSURF.fill(WHITE)
    P1.draw(DISPLAYSURF)
    E1.draw(DISPLAYSURF)
    for coin in coins:
        coin.draw(DISPLAYSURF)

    # Отображение счета
    score_text = font.render(f"Coins: {score}", True, BLACK)
    DISPLAYSURF.blit(score_text, (SCREEN_WIDTH - 120, 10))

    # Проверка на столкновение с противником
    if P1.rect.colliderect(E1.rect):
        print("Game Over")
        pygame.quit()
        sys.exit()

    pygame.display.update()
    FramePerSec.tick(FPS)
```
