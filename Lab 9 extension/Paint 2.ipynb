```python
import pygame
import sys

# Инициализация Pygame
pygame.init()

# Константы
WIDTH, HEIGHT = 800, 600
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
RED = (255, 0, 0)
BLUE = (0, 0, 255)
GREEN = (0, 255, 0)

# Настройки экрана
screen = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("Paint Tool")
screen.fill(WHITE)

# Переменные
clock = pygame.time.Clock()
drawing = False
last_pos = None
mode = "pen"  # Режимы: 'pen', 'rect', 'square', 'circle', 'right_triangle', 'equilateral_triangle', 'rhombus', 'eraser'
color = BLACK
radius = 10  # Размер кисти или ластика

# Основной цикл
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

        elif event.type == pygame.KEYDOWN:
            if event.key == pygame.K_r:
                mode = "rect"  # Рисование прямоугольников
            elif event.key == pygame.K_s:
                mode = "square"  # Рисование квадратов
            elif event.key == pygame.K_c:
                mode = "circle"  # Рисование кругов
            elif event.key == pygame.K_t:
                mode = "right_triangle"  # Прямоугольный треугольник
            elif event.key == pygame.K_e:
                mode = "equilateral_triangle"  # Равносторонний треугольник
            elif event.key == pygame.K_h:
                mode = "rhombus"  # Ромб
            elif event.key == pygame.K_p:
                mode = "pen"  # Режим кисти
            elif event.key == pygame.K_x:
                mode = "eraser"  # Ластик
            elif event.key == pygame.K_1:
                color = BLACK  # Черный цвет
            elif event.key == pygame.K_2:
                color = RED  # Красный цвет
            elif event.key == pygame.K_3:
                color = GREEN  # Зеленый цвет
            elif event.key == pygame.K_4:
                color = BLUE  # Синий цвет

        elif event.type == pygame.MOUSEBUTTONDOWN:
            drawing = True
            last_pos = event.pos

        elif event.type == pygame.MOUSEBUTTONUP:
            drawing = False
            x1, y1 = last_pos
            x2, y2 = event.pos

            if mode == "rect":  # Прямоугольник
                pygame.draw.rect(screen, color, pygame.Rect(min(x1, x2), min(y1, y2), abs(x2 - x1), abs(y2 - y1)), 2)
            elif mode == "square":  # Квадрат (размер по наименьшей стороне)
                side = min(abs(x2 - x1), abs(y2 - y1))
                pygame.draw.rect(screen, color, pygame.Rect(x1, y1, side, side), 2)
            elif mode == "circle":  # Круг
                radius = max(abs(x2 - x1), abs(y2 - y1)) // 2
                pygame.draw.circle(screen, color, last_pos, radius, 2)
            elif mode == "right_triangle":  # Прямоугольный треугольник
                pygame.draw.polygon(screen, color, [(x1, y1), (x2, y2), (x1, y2)], 2)
            elif mode == "equilateral_triangle":  # Равносторонний треугольник
                height = abs(y2 - y1)
                half_base = height // 2
                pygame.draw.polygon(screen, color, [(x1, y2), (x1 + half_base, y1), (x1 - half_base, y1)], 2)
            elif mode == "rhombus":  # Ромб
                width = abs(x2 - x1)
                height = abs(y2 - y1)
                pygame.draw.polygon(screen, color, [(x1, y1 + height // 2), (x1 + width // 2, y1), 
                                                     (x1, y1 - height // 2), (x1 - width // 2, y1)], 2)

        elif event.type == pygame.MOUSEMOTION and drawing:
            if mode == "pen":  # Рисование кистью
                pygame.draw.line(screen, color, last_pos, event.pos, radius)
                last_pos = event.pos
            elif mode == "eraser":  # Ластик
                pygame.draw.line(screen, WHITE, last_pos, event.pos, radius)
                last_pos = event.pos

    pygame.display.flip()
    clock.tick(60)

pygame.quit()
sys.exit()
```

 Равносторонний треугольник (Клавиша E)
 
 Ромб (Клавиша H)
 
 Кисть (P) и ластик (X)

1 – Черный
